## Multivariate Linear Regression for Diabetes dataset

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection, metrics
```

Load the diabetes dataset — all 10 features

```python
X, y = datasets.load_diabetes(return_X_y=True)
print('X shape:', X.shape)
print('y shape:', y.shape)
```

Split into training and test sets

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33, random_state=42)
```

Fit the multivariate linear regression model

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
```

Predict and evaluate

```python
y_pred = model.predict(X_test)
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('R² Score:', metrics.r2_score(y_test, y_pred))
```

Plot predicted vs actual values

```python
plt.scatter(y_test, y_pred, color='black')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='blue', linewidth=3)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted - Multivariate Linear Regression')
plt.show()
```
