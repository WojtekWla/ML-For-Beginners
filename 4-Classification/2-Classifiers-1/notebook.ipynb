# Build Classification Models

```python
import pandas as pd
cuisines_df = pd.read_csv("../data/cleaned_cuisines.csv")
cuisines_df.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, classification_report, precision_recall_curve
from sklearn.svm import SVC
import numpy as np 
```

```python
cuisines_label_df = cuisines_df["cuisine"]
cuisines_label_df.head()
```

```python
cuisined_feature_df = cuisines_df.drop(["cuisine", "Unnamed: 0"], axis=1)
cuisined_feature_df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(cuisined_feature_df, cuisines_label_df, test_size=0.3, random_state=42)
```

```python
lr = LogisticRegression(solver="lbfgs")
model = lr.fit(X_train, y_train)

accuracy = model.score(X_test, y_test)
print(f"accuracy {accuracy}")
```

```python
print(f'ingredients: {X_test.iloc[50][X_test.iloc[50]!=0].keys()}')
print(f'cuisine: {y_test.iloc[50]}')
```

```python
test = X_test.iloc[50].values.reshape(-1, 1).T
proba = model.predict_proba(test)
classes = model.classes_

resultdf = pd.DataFrame(data=proba, columns=classes)
topPredictions = resultdf.T.sort_values(by=[0], ascending=[False])
topPredictions.head()
```

```python
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```
